```python
from pathlib import Path
import json
import pandas as pd
import numpy as np

PROJECT_ROOT = Path.cwd()

if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

artifacts_dir = PROJECT_ROOT / "artifacts"
raw_data_path = PROJECT_ROOT / "data" / "raw" / "nasa_exoplanet_intelligence.csv"

nasa_df = pd.read_csv(raw_data_path)

with open(artifacts_dir / "ui_feature_config.json", "r", encoding="utf-8") as f:
    ui_config = json.load(f)

# Unknown sınıfını UI rehberi için kullanmayalım
guide_df = nasa_df[nasa_df["planet_type"] != "Unknown"].copy()

# Slider scale kararları
log_scale_features = [
    "planet_mass_earth",
    "orbital_period_days",
    "semi_major_axis_au",
    "dist_from_earth_pc"
]

linear_scale_features = [
    "n_stars",
    "n_planets",
    "planet_radius_earth",
    "equilibrium_temp_k",
    "orbital_eccentricity",
    "star_temp_k",
    "star_radius_sun",
    "star_mass_sun",
    "star_age_gyr",
    "star_surface_gravity",
    "star_metallicity"
]

for feature in ui_config.keys():
    if feature in log_scale_features:
        ui_config[feature]["scale"] = "log"
    else:
        ui_config[feature]["scale"] = "linear"

# Sınıf marker'ları: her feature için planet_type median değerleri
for feature in ui_config.keys():
    if feature not in guide_df.columns:
        continue

    markers = (
        guide_df
        .groupby("planet_type")[feature]
        .median()
        .dropna()
        .round(4)
        .to_dict()
    )

    ui_config[feature]["class_markers"] = markers

# Daha kullanışlı slider aralıkları
range_overrides = {
    "n_stars": {
        "min": int(guide_df["n_stars"].min()),
        "max": int(guide_df["n_stars"].max()),
        "step": 1
    },
    "n_planets": {
        "min": int(guide_df["n_planets"].min()),
        "max": int(guide_df["n_planets"].max()),
        "step": 1
    },
    "orbital_period_days": {
        "min": 0.09,
        "max": 2500.0,
        "step": 0.1
    },
    "planet_radius_earth": {
        "min": 0.3,
        "max": 18.0,
        "step": 0.01
    },
    "planet_mass_earth": {
        "min": 0.02,
        "max": 10000.0,
        "step": 0.01
    },
    "equilibrium_temp_k": {
        "min": 34,
        "max": 4050,
        "step": 1
    },
    "orbital_eccentricity": {
        "min": 0,
        "max": 0.95,
        "step": 0.01
    },
    "semi_major_axis_au": {
        "min": 0.005,
        "max": 5.0,
        "step": 0.001
    },
    "star_temp_k": {
        "min": 3500,
        "max": 8000,
        "step": 1
    },
    "star_radius_sun": {
        "min": 0.01,
        "max": 10.0,
        "step": 0.01
    },
    "star_mass_sun": {
        "min": 0.01,
        "max": 3.0,
        "step": 0.01
    },
    "star_age_gyr": {
        "min": 0,
        "max": 16.5,
        "step": 0.1
    },
    "star_surface_gravity": {
        "min": 0.5,
        "max": 8.1,
        "step": 0.01
    },
    "star_metallicity": {
        "min": -1.0,
        "max": 0.6,
        "step": 0.01
    },
    "dist_from_earth_pc": {
        "min": 1,
        "max": 10000,
        "step": 1
    }
}

for feature, rules in range_overrides.items():
    if feature in ui_config:
        ui_config[feature].update(rules)

with open(artifacts_dir / "ui_feature_config.json", "w", encoding="utf-8") as f:
    json.dump(ui_config, f, ensure_ascii=False, indent=4)

print("Updated:", artifacts_dir / "ui_feature_config.json")
```
